## Exibição do primeiro Gráfico com Matplotlib para o Top 10 Principais Locais com a taxa mais alta de Homicídios Intecionais

```python
top_rate = df.sort_values(by = 'Rate', ascending = False).head(10)

plt.figure(figsize = (14, 8))
plt.style.use('dark_background')
colorpalette = sns.color_palette('autumn')

plt.bar(top_rate['Location'], top_rate['Rate'], color=(colorpalette), edgecolor='white')
plt.xlabel('Locação', fontsize=14, fontweight='bold', style='italic')
plt.ylabel('Contagem')
plt.title('Os 10 Principais Locais com a Taxa mais Alta de Homicídios Intecionais\n', fontsize = 20, fontweight='bold', style='italic')
plt.xticks(rotation = 65, fontweight='bold', style='italic')
for x,y in zip(top_rate['Location'],top_rate['Rate']):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=11, fontweight='bold', style='italic')
    
plt.show()
```

## Execução de Gráfico referente as Locações ( Países ) com maior Número de Homicidios Intencional

```python
top_count = df.sort_values(by = 'Count', ascending = False).head(10)

plt.figure(figsize = (14, 8))
plt.style.use('dark_background')

plt.bar(top_count['Location'], top_count['Count'], color=(colorpalette), edgecolor='white')
plt.xlabel('\nPaíses', fontweight='bold', style='italic')
plt.ylabel('Contagem')
plt.title('\nTop 10 Países com Contagem mais Alta de Homicidios Intencionais\n', fontsize = 16, fontweight='bold', style='italic')
plt.xticks(rotation=40, fontweight='bold', style='italic')
for x,y in zip(top_count['Location'],top_count['Count']):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=11, fontweight='bold', style='italic')

plt.show()
```

##  Visualizando em Gráficos Informações Precisas dos Homicidios por Região, Ano e SubRegião
### Executando uma Função para Criação dos Gráficos em Diferentes Estilos e com Diferentes Refencias do Arquivo com Colunas e Linhas

```python
def plots(df, x, y):
    plt.style.use('dark_background')
    f, ax = plt.subplots(1, 2, figsize = (20, 10))
    Group_data = df.groupby(x)
    sns.barplot(x = Group_data[y].mean().index, y = Group_data[y].mean().values, ax = ax[0], palette = 'autumn', edgecolor='white')
    
    for container in ax[0].containers:
        ax[0].bar_label(container, color = 'white', size = 17)

    palette_color = sns.color_palette('winter')
    plt.pie(x = df[x].value_counts(),
            labels = df[x].value_counts().index,
            autopct = '%.0f%%',
            shadow = True,
            colors = palette_color)
    plt.suptitle(x, fontsize = 40)
    plt.show()
```

```python
plots(df, 'Region', 'Rate')
plots(df, 'Year', 'Rate')
```

# Analisando os Top 15 Primeiros SubRegioes com maior Percentual de Homicídios

```python
df_subregion = df['Subregion'].value_counts().nlargest(10)

palette___color = sns.color_palette('winter')
plt.figure(figsize = (14, 8))
plt.style.use('dark_background')

plt.pie(x = df_subregion, labels = df_subregion.index, colors=(palette___color), shadow=True, autopct='%1.1f%%')
plt.title('\nTop 10 SubRegiões com Percentual mais Alto de Homicidios Intencionais\n', fontsize = 16, fontweight='bold', style='italic')


plt.show()
```

## Verificando a Evolução do Número de Homicídios ao Longo dos Anos

```python
year_count = df['Year'].value_counts().nlargest(20)

plt.figure(figsize = (11, 9))
plt.title('Evolução do Número de Homicídios ao Longo dos Anos\n', fontsize = 17, fontweight='bold', style='italic')
sns.barplot(x=year_count.values, y=year_count.index, orient='h', palette = 'winter', edgecolor='white')
plt.xticks(fontweight='bold', style='italic')
plt.yticks(fontweight='bold', style='italic')

plt.show()
```
