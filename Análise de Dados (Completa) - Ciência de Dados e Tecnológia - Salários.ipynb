# Análise Dados Referênte a Salários na Área de Ciências de Dados e T.I


## Introdução :
### A Análise de Salários é um Processo que envolve a Coleta, Organização e Interpretação dos Dados Relativos à Remuneração dos Profissionais de uma Determinada Área da Tecnológia da Informação, focando em específico ao financeiro. Aqui será Apresentada uma Abordagem para Realizar essa Análise Utilizando Ferramentas da área de Ciência de Dados e T.I (Tecnologias de Informação), como o Matplotlib PyPlot, Seaborn e outros. Além disso, serão discutidas as principais etapas necessárias para realizar tal Análise, bem como os tipos de Informações que devem ser Coletadas e Analisadas.
## Etapas da Análise de Salários :

### 1. Coleta de Dados: O primeiro passo da análise de salários é a coleta de dados relevantes. Nessa etapa, são obtidos todos os registros de Salário da Empresa ou instituição financeira, incluindo o nome do funcionário, cargo, data de entrada no trabalho, salário bruto, descontos, bonificação, etc.

### 2. Organização de Dados: Após a coleta dos dados, eles precisam ser organizados de forma adequada. Isso inclui a criação de bancos de dados, tabelas e campos que armazenar todos os dados coletados.

### 3. Limpeza de Dados: Na limpeza de dados, é feito o tratamento de valores ausentes, erros, inconsistências e outras irregularidades nos dados.

### 4. Descriptive Statistics: Com a descriptiva estatística, é possível obter informações gerais sobre os dados, como média, mediana, moda, variância, desvio padrão, entre outros.

### 5. Visualização de Dados: A visualização de dados permite representar os dados de maneira mais facilmente entendida pelos usuários. Exemplos de gráficos que podem ser utilizados nesse processo incluem histogramas, barras horizontais, caixas de diálogo, dispersões, linhas temporais, entre outros.

### 6. Modelagem de dados: A modelagem de dados consiste em criar algoritmos e modelos para prever resultados com base nas informações coletadas. Essa etapa pode ser utilizada para predizer o valor futuro do salário, por exemplo.

### 7. Interpretação de Resultados: Finalmente, após a realização da análise, é necessário interpretarem os resultados e tomar decisões baseadas nestes dados.

## Análise de Salário Médio por Ano

```python
plt.style.use('dark_background')

# Compreendendo a Distribuição Salarial
mean_salary = df.groupby('work_year')['salary_in_usd'].mean()

# Criando Subparcelas com Eixo x Compartilhado
fig = make_subplots()

# Adicionando Traço de Barra
bar_trace = go.Bar(
    x=mean_salary.index,
    y=mean_salary.values,
    name="Mean Salary",
    marker=dict(color=cor_indivi_1) 
)
fig.add_trace(bar_trace)

# Adicionando Traço de Linha
line_trace = go.Scatter(
    x=mean_salary.index,
    y=mean_salary.values,
    name="Mean Salary",
    mode="lines+markers", 
    line=dict(color='black'),
    marker=dict(symbol='circle-open', size=10)
)
fig.add_trace(line_trace)

# Atualizando layout
fig.update_layout(
    title="Analisando o Salário Médio por Ano",
    xaxis_title="Trabalho por Ano",
    yaxis_title="Mádia Salarial",
    height=600
)

fig.show()
```
